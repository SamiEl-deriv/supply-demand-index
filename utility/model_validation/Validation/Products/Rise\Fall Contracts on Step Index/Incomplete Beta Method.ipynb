```python
from scipy.special import betainc
import yaml
import pandas as pd
```

```python
# Define the file path to the YAML file
yaml_file_path = 'steprng_rise_fall_theo_probability.yaml'

# Open and read the YAML file
with open(yaml_file_path, 'r') as file:
    config = yaml.safe_load(file)

# Print the content of the YAML file
# display(config)
```

```python
def get_prob(n, allow_equal):
    p = 0.5
    if n % 2: #odd
        k = (n+1)/2 # k greater
        greater = betainc(k, n - k + 1, p)
        greater_equal = greater
        equal = 0 
    else: #even
        k = n/2
        greater = betainc(k+1, n - (k+1) + 1, p)
        greater_equal = betainc(k, n - (k) + 1, p)
        equal = greater_equal - greater
    smaller = 1 - greater_equal

    if allow_equal: 
        return (round(greater_equal,4), round(smaller + equal,4))
    else:
        return (round(greater,4), round(smaller,4))
    
```

```python
df = pd.DataFrame(columns = ['ticks','yaml_rise_fall_equal_prob','calc_rise_fall_equal_prob',\
                            'yaml_rise_fall_prob','calc_rise_fall_prob','pass'])
```

```python
df = pd.DataFrame(columns = ['ticks','yaml_rise_fall_equal_prob','calc_rise_fall_equal_prob',\
                            'yaml_rise_fall_prob','calc_rise_fall_prob','pass'])
for tick in config.keys():
    u_t,d_t = get_prob(tick, True)
    u_f,d_f = get_prob(tick, False)
    pass_f = True
    if not ((list([d_t,u_t]) == list(config[tick]['allow_equals_true'].values())) &\
            (list([d_f,u_f]) == list(config[tick]['allow_equals_false'].values()))):
        pass_f = False
        
    df.loc[len(df)] = tick, list(config[tick]['allow_equals_true'].values()), list([d_t,u_t]),\
                        list(config[tick]['allow_equals_false'].values()), list([d_f,u_f]), pass_f\
                    
print('Pass cases:',len(df[df['pass']==True]))
df
```
