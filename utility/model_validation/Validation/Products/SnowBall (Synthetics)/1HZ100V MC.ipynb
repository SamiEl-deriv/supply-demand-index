```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numba import njit
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
pd.set_option('display.max_colwidth',300)
import random
import math
```

```python
@njit
def simulate_stock_price(s, dt, v, ticks):
    results = [s]
    for j in range(ticks):
        results.append(round(results[-1] * math.exp(-v**2/2 * dt + v * math.sqrt(dt) * random.normalvariate(0, 1)),2))
#         results.append(round(results[-1] * math.exp(-v**2/2 * dt + v * math.sqrt(dt) * np.random.normal(0, 1)),2))
    return results
```

```python
paths = []
for i in range(2000):
    path = simulate_stock_price(450, 1/365/86400, 1, 86400*30)
    paths.append([path[0], path[int(len(path)/2)],path[-1]])
#     plt.plot([path[0], path[int(len(path)/2)],path[-1]], label=f'Line {i+1}')
for i, path in enumerate(paths):
    plt.plot(path, label=f'Line {i+1}')
# Add labels and title
plt.xlabel('Ticks')
plt.ylabel('Spot')
plt.title('1HZ100V')
# plt.legend()

# Show the plot
plt.grid()
plt.show()
```
